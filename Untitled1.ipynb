# Drop the Sex column

```python
dataset  = dataset.drop(columns= 'Sex')
```

# Split the dataset

```python
features = [x for x in dataset.columns if x not in 'Class_number_of_rings']
```

```python
X_train , _x, y_train , _y = train_test_split(dataset[features],dataset['Class_number_of_rings'],train_size = 0.6)

X_val,X_test , y_val,y_test  = train_test_split(_x , _y ,train_size = 0.5)

del _x,_y
```

```python
fig, ax  = plt.subplots(figsize= (12,3))
X_train.plot(x = 'Length',y = 'Diameter',kind = 'scatter', ax = ax , label = 'train set')
X_val.plot(x = 'Length',y = 'Diameter',kind = 'scatter', ax = ax, color = 'r' , label = 'evaluate set')
X_test.plot(x = 'Length',y = 'Diameter',kind = 'scatter', ax = ax, color = 'g' , label = 'test set')
```

# XGboost

## train model

```python
xgb = XGBRegressor(n_estimators = 500, learning_rate = 0.1,verbosity = 1)
xgb.fit(X_train,y_train, eval_set = [(X_val,y_val)])
```

## best itteration

```python
xgb.best_iteration
```

## Evaluate model

```python
print(f"Training MSE: {mean_squared_error(xgb.predict(X_train),y_train)/2 :.2f}")
print(f"Cross Validation MSE: {mean_squared_error(xgb.predict(X_val),y_val)/2 :.2f}")
print(f"Test MSE: {mean_squared_error(xgb.predict(X_test),y_test)/2 :.2f}")
```

Note : variance

### avoid overfiting

try smaller set of estimators

```python
xgb = XGBRegressor(n_estimators = 500, learning_rate = 0.1,verbosity = 1)
xgb.fit(X_train,y_train, eval_set = [(X_val,y_val)], early_stopping_rounds = 10)
```

```python
print(f"Training MSE: {mean_squared_error(xgb.predict(X_train),y_train)/2 :.2f}")
print(f"Cross Validation MSE: {mean_squared_error(xgb.predict(X_val),y_val)/2 :.2f}")
print(f"Test MSE: {mean_squared_error(xgb.predict(X_test),y_test)/2 :.2f}")
```

## avoid underfitting

```python
poly = PolynomialFeatures(degree = 2,include_bias=False)
X_train_poly = poly.fit_transform(X_train)
X_val_poly = poly.fit_transform(X_val)
X_test_poly = poly.fit_transform(X_test)
st  = StandardScaler()
X_train_poly   = st.fit_transform(X_train_poly)
X_val_poly   = st.fit_transform(X_val_poly)
X_test_poly   = st.fit_transform(X_test_poly)
```

```python
li = LinearRegression()
li.fit(X_train_poly,y_train)
```

```python
print(f"Training MSE: {mean_squared_error(li.predict(X_train_poly ),y_train)/2 :.2f}")
print(f"Cross Validation MSE: {mean_squared_error(li.predict(X_val_poly ),y_val)/2 :.2f}")
print(f"Test MSE: {mean_squared_error(li.predict(X_test_poly ),y_test)/2 :.2f}")
```

# Neural Network

```python
model = Sequential(
   [Dense(40, activation = 'relu'),
            Dense(15, activation = 'relu'),
            Dense(1, activation = 'relu')]
)

model.compile(optimizer = tf.keras.optimizers.Adam(0.01),
              loss = tf.keras.losses.BinaryCrossentropy())
```

```python
model.fit(X_train,y_train,epochs=300,
        verbose=1)
```

##Evaluate model

```python
print(f"Training MSE: {mean_squared_error(model.predict(X_train),y_train)/2 :.2f}")
print(f"Cross Validation MSE: {mean_squared_error(model.predict(X_val),y_val)/2 :.2f}")
print(f"Test MSE: {mean_squared_error(model.predict(X_test),y_test)/2 :.2f}")
```
